```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv(r'C:\Users\anupa\Downloads\Fake news Dataset\train.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
train_data = train_data.fillna(' ')
transformer = TfidfTransformer(smooth_idf=False)
count_vectorizer = CountVectorizer(ngram_range=(1, 2))
counts = count_vectorizer.fit_transform(train_data['title'].values)
X = transformer.fit_transform(counts)
Y = train_data['label'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np
import itertools
```

MULTINOMIAL NAIVE BAYES WITHOUT HYPERPARAMTER

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train,y_train)

pred = classifier.predict(X_train)
accuracy = metrics.accuracy_score(y_train, pred)
print("Train accuracy:   %0.3f" % accuracy)

pred = classifier.predict(X_test)
accuracy = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % accuracy)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
labels = ['NOT REAL', 'REAL']
disp = plot_confusion_matrix(classifier, X_test, y_test,
                                 display_labels = labels,
                                 cmap = plt.cm.Blues,
                                 normalize = 'all')
title = "MULTINOMIAL NAIVE BAYES"
disp.ax_.set_title(title)
print(title)
print(disp.confusion_matrix)
```

STACKING CLASSIFIER

```python
from sklearn.ensemble import StackingClassifier
```

```python
classifier_2 = RandomForestClassifier()
classifier_3 = AdaBoostClassifier(n_estimators = 6, learning_rate = 0.1)
classifier_4 = LogisticRegression()

estimators = [('rf', classifier_2),
              ('ada', classifier_3),
              ('logis', classifier_4)
             ]

classifier = StackingClassifier(estimators = estimators, final_estimator = SVC())
classifier.fit(X_train, y_train)

pred = classifier.predict(X_train)
accuracy = metrics.accuracy_score(y_train, pred)
print("Train accuracy:   %0.3f" % accuracy)

pred = classifier.predict(X_test)
accuracy = metrics.accuracy_score(y_test, pred)
print("Test accuracy:   %0.3f" % accuracy)

labels = ['NOT REAL', 'REAL']
disp = plot_confusion_matrix(classifier, X_test, y_test,
                                 display_labels = labels,
                                 cmap = plt.cm.Blues,
                                 normalize = 'all')
title = "STACKING CLASSIFIER"
disp.ax_.set_title(title)
print(title)
print(disp.confusion_matrix)
```
